```python
df_feedback_data = pd.read_csv('drive/MyDrive/NBA_Recap_Summary_Generator/game_recaps_with_rewritten_summaries_improved.csv')
print(df_feedback_data.columns)
df_feedback_data.head()
```

```python
NBA_TEAMS = {
    "ATL": "Hawks",
    "BRK": "Nets",
    "BOS": "Celtics",
    "CHO": "Hornets",
    "CHI": "Bulls",
    "CLE": "Cavaliers",
    "DAL": "Mavericks",
    "DEN": "Nuggets",
    "DET": "Pistons",
    "GSW": "Warriors",
    "HOU": "Rockets",
    "IND": "Pacers",
    "LAC": "Clippers",
    "LAL": "Lakers",
    "MEM": "Grizzlies",
    "MIA": "Heat",
    "MIL": "Bucks",
    "MIN": "Timberwolves",
    "NOP": "Pelicans",
    "NYK": "Knicks",
    "OKC": "Thunder",
    "ORL": "Magic",
    "PHI": "76ers",
    "PHO": "Suns",
    "POR": "Blazers",
    "SAC": "Kings",
    "SAS": "Spurs",
    "TOR": "Raptors",
    "UTA": "Jazz",
    "WAS": "Wizards"
}
df_feedback_data['home_team'] = df_feedback_data['home_team'].map(NBA_TEAMS)
df_feedback_data['away_team'] = df_feedback_data['away_team'].map(NBA_TEAMS)
df_feedback_data.head()
```

```python
df_feedback_data['wrong_team_mentioned'] = df_feedback_data[['home_team', 'away_team', 'rewritten_summary']].apply(lambda x: x['home_team'] not in x['rewritten_summary'] or x['away_team'] not in x['rewritten_summary'], axis=1)
df_feedback_data.head()
```

```python
relevant_columns = ['query', 'game_recap', 'game_recap_summary_ground_truth', 'game_recap_summary_generated', 'rewritten_summary', 'wrong_team_mentioned']
df_feedback_data = df_feedback_data[relevant_columns]
df_feedback_data.head()
```
